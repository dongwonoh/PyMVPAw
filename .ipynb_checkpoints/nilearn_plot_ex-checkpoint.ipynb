### Brain plotting in python

New RyMVPA functions to plot output data from RyMVPA (via sl2nifti and datadict2nift). I have yet to sort out how to plot directly from PyMVPA datasets, there seem to be some minor incompatibilities that are difficult to unearth. So, going with this for now. 

**Important note**: Currently only thresholds to show values greater than a set threshold.

Goal was to make plots from [MIND summer school here](https://github.com/Summer-MIND/mind_2017/blob/master/Tutorials/Decoding/Chang_Decoding.ipynb).

Some doc:
- http://nilearn.github.io/manipulating_images/input_output.html
- http://nilearn.github.io/auto_examples/plot_3d_and_4d_niimg.html
- http://nilearn.github.io/plotting/index.html
- https://github.com/ljchang/nltools
